```python
from bs4 import BeautifulSoup # HTMLやXMLの文書を解析し、データを抽出するためのBeautiful Soupライブラリをインポート（通常、'bs4'と省略される）
import requests # 外部のウェブサイトにHTTPリクエストを送信するためのライブラリをインポート
from urllib.parse import urljoin, urlparse # URLの操作に関する関数を提供するurllib.parseモジュールからurljoinとurlparseをインポート
import time # timeモジュールをインポート（時間関連の関数を提供）
```

```python
visited = set() # 訪問済みのURLを記録するためのセット
title_dict = {} # URLとタイトルを対応付けるための辞書

def crawl(url): # 指定したURLをクロールする関数
    if url in visited: # すでに訪問済みのURLであれば
        return # 処理を終了

    visited.add(url) # 訪問済みのURLに追加

    try: # URLにアクセスしてレスポンスを取得
        res = requests.get(url) 
    except Exception as e: # アクセスエラーが発生した場合の例外処理
        print("アクセスエラー:", url, e)
        return

    soup = BeautifulSoup(res.text, "html.parser") # レスポンスの内容を解析してBeautifulSoupオブジェクトを作成

    # タイトル取得（なければ空）
    title = soup.title.get_text(strip=True) if soup.title else ""
    title_dict[url] = title

    # サイトに負荷を与えないように1秒休憩
    time.sleep(1)

    # 同一ドメインのリンクを再帰的に巡回
    for a in soup.find_all("a", href=True):
        link = urljoin(url, a["href"])
        if urlparse(link).netloc == urlparse(url).netloc:
            crawl(link)
```
